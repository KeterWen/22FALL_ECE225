```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing

plt.style.use('ggplot')
```

```python
def ridge(X, y, lamb):
    """
    Ridge Regression Solution
    """
    inv = np.linalg.inv(np.dot(np.transpose(X), X) + lamb * np.eye(X.shape[1]))
    matrix = np.dot(inv, np.transpose(X))
    return np.dot(matrix, y)
```

```python
data = pd.read_csv("insurance.csv")
```

```python
X = data[["age", "bmi", "children"]].to_numpy()
y = data[["charges"]].to_numpy()
```

```python
for i in range(X.shape[1]):
    X[:,i] = preprocessing.scale(X[:,i])
```

```python
w1 = []
w2 = []
w3 = []
x = []

for i in range(-4, 16):
    x.append(2 ** i)
    w = ridge(X, y, 2 ** i)
    w1.append(w[0])
    w2.append(w[1])
    w3.append(w[2])
```

```python
plt.plot(x, w1, color = 'r', label="w_1")
plt.plot(x, w2, color = 'g', label="w_2")
plt.plot(x, w3, color = 'b', label="w_3")
plt.xlabel("lambda")
plt.ylabel("weight")
plt.legend(loc = "best")
plt.savefig("Ridge_Regression")
```
